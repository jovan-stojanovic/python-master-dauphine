## Python: qu'est ce que c'est?

Python est un langage de programmation créé en 1991 par Guido van Rossum. C'est un langage libre et gratuit, développé par une large communauté d'utilisateurs.  

Pour résumer, le succès de Python tient particulièrement en quatre points:
* **1. Simplicité et flexibilité:** le langage est simple (à lire) à comprendre et la librairie standard propose de nombreuses fonctionalités par défaut.
* **2. Des librairies spécialisées dans tous les domaines:** des extensions du langage Python pour réaliser de nombreuses tâches:
    - `Django`: framwork de developpement web.
    - `Numpy`: librairie d'algébre linéaire.
    - `scikit-learn`: librairie dédiée à l'aprentissage automatique.
    - `Matplotlib`: librairie pour visualisation de données en 2-D.
    - ....

  Ces librairies sont réunies dans 2 principaux gestionnaires de librairies: `pip` et `conda` (comme `CRAN` pour les librairies `R`).
* **3. Grande communauté d'utilisateurs:** utilisation, documentation et entretien des librairies open-source qui rende l'ecosystème de Python particulièment riche et efficace.
* **4. Une facilité d'extension:** Python est utilisable sur Windows, Linux/Unix, MacOS. Il permet aussi l'utilisation d'autres langages dans un programme Python, par exemple `R` avec `rpy2`.

<b style="font-variant: small-caps; text-decoration: underline;">Langage Interprété:</b><br>
    Python est un `langage interprété', c'est à dire qu'il se compose d'un `langage' et d'un `interpréteur'.
    La principale différence avec les `langages compilés' est que la suite (l'ensemble) des instructions qui va être exécutée par le programme n'est pas connue quand celui-ci est lancé.
    Le langage Python définit un ensemble d'instructions à partir de la ligne de code fournie.<br>
    Il existe ensuite différents interpréteurs qui prennent ces instructions pour modifier l'état interne (comprendre les variables) du programme:

<b style="font-variant: small-caps; text-decoration: underline;">Orienté Objet:</b><br>
    Python est un langage `orienté objet'. Le langage se base sur la définitions d'un ensemble d'objets tel que les listes (list), les chaines de charactères (string) ou les numériques (int ou float). Tout ce qu'on manipule sera, au final, un objet. Il sera également possible de définir soi-même un nouvel objet.

### Tuples

 * Les *tuples* (n-uplets) ressemblent aux listes mais ils sont *immuables* : ils ne peuvent pas être modifiés une fois créés.
 
 * On les crée avec la syntaxe `(..., ..., ...)` ou simplement `..., ...`:


````python
point = (10, 20)
print(point, type(point))
````

````python
point[0]
````

Un *tuple* peut être dépilé par assignation à une liste de variables séparées par des virgules :

````python
x, y = point

print("x =", x)
print("y =", y)
````

On ne peut pas faire :

````python
point[0] = 20
````

## Conditions, branchements et boucles

### Branchements: if, elif, else

````python
a = 3
b = 2

if a < b:  # notez the : et l'indentation
    print("a est strictement inférieur à b")
elif a > b:
    print("b est strictement inférieur à a")
else:
    print("b est égal à a")
````

## Boucles

Boucles **`for`**:

````python
for x in [1, 2, 3]:
    print(x)
````

La boucle `for` itère sur les éléments de la list fournie. Par exemple:

````python
for x in range(4): # par défault range commence à 0
    print(x)
````

## Classes

 * Les *classes* sont les éléments centraux de la *programmation orientée objet*

 * Classe: structure qui sert à représenter un objet et l'ensemble des opérations qui peuvent êtres effectuées sur ce dernier.

Dans Python une classe contient des *attributs* (variables) et des *méthodes* (fonctions). Elle est définie de manière analogue aux fonctions mais en utilisant le mot clé `class`. La définition d'une classe contient généralement un certain nombre de méthodes de classe (des fonctions dans la classe).

* Le premier argument d'un méthode doit être `self`: argument obligatoire. Cet objet `self` est une auto-référence.
* Certains noms de méthodes ont un sens particulier, par exemple : 
   * `__init__`: nom de la méthode invoquée à la création de l'objet
   * `__str__` : méthode invoquée lorsque une représentation de la classe sous forme de chaîne de caractères est demandée, par exemple quand la classe est passée à `print`
   * voir http://docs.python.org/2/reference/datamodel.html#special-method-names pour les autres noms spéciaux

### Exemple
  

````python
class Point(object):
    """
    Classe pour représenter un point dans le plan.
    """
    def __init__(self, x, y):
        """
        Creation d'un nouveau point en position x, y.
        """
        self.x = x
        self.y = y
        
    def translate(self, dx, dy):
        """
        Translate le point de dx and dy.
        """
        self.x += dx
        self.y += dy
        
    def __str__(self):
        return f"Point: [{self.x}, {self.y}]"
````

Pour créer une nouvelle instance de la classe:

````python
p1 = Point(x=0, y=0) # appel à __init__
print(p1.x)
print(p1.y)
print(p1)         # appel à la méthode __str__
````

````python
p1.translate(dx=1, dy=1)
print(p1)
print(type(p1))
````

Pour invoquer une méthode de la classe sur une instance `p` de celle-ci:

````python
p2 = Point(1, 1)

p1.translate(0.25, 1.5)

print(p1)
print(p2)
````

### Remarques

 * L'appel d'une méthode de classe peut modifier l'état d'une instance particulière
 * Cela n'affecte ni les autres instances ni les variables globales

## Exceptions

 * Dans Python les erreurs sont gérées à travers des *"Exceptions"*
 * Une erreur provoque une *Exception* qui interrompt l'exécution normale du programme
 * L'exécution peut éventuellement reprendre à l'intérieur d'un bloc de code `try` - `except`


* Une utilisation typique: arrêter l'exécution d'une fonction en cas d'erreur:

        def my_function(arguments):
    
        if not verify(arguments):
            raise Expection("Invalid arguments")
        
        # et on continue

On utilise `try` et  `expect` pour maîtriser les erreurs:

    try:
        # normal code goes here
    except:
        # code for error handling goes here
        # this code is not executed unless the code
        # above generated an error

Par exemple:

````python
try:
    print("test_var")
    # genere une erreur: la variable test n'est pas définie
    print(test_var)
except:
    print("Caught an expection")
````

Pour obtenir de l'information sur l'erreur: accéder à l'instance de la classe `Exception` concernée:

    except Exception as e:

````python
try:
    print("test")
    # generate an error: the variable test is not defined
    print(test)
except Exception as e:
    print("Caught an expection:", e)
````
